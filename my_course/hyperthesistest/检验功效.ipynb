```python
%%manim --disable_caching -qh Power
from con import *
from my_course.my_func import *
from scipy.stats import norm
from math import sqrt
class Power(Scene):
    config['tex_template'] = TexTemplateLibrary.ctex
    def multi_play(self,ani,wait_time=False,tim=None,time_value=1):
        if tim == None:
            tim = [time_value]*len(ani)
        for i in range(len(ani)):
            if ani[i][0]=='a':
                self.add(*ani[i][1:])
            elif ani[i][0]=='r':
                self.remove(*ani[i][1:])
            elif ani[i][0]=='m':
                ani[i][1:]
            else:
                self.play(*ani[i][:])
            
            if wait_time:
                if tim[i]>0:
                    self.wait(tim[i]) 
            else:
                pass   
    def change_time(self,length,change_list):
        time_list = [0]*length
        for i in change_list:
            time_list[i] = 1
        return time_list
    def type_error(self):
        title1 = Text('III',color=c_blue).to_edge(UL)
        title2 = Tex(r'$\beta\mbox{}$',font_size=48,color=c_blue).to_edge(UL)
        type1 = Tex(r'$\mbox{I}\enspace\mbox{}$')
        type1.next_dl(title1[2],buff=type1.get_height()+0.5)
        type2 = Tex('II').next_dl(type1)
        tx2 = Tex('').next_to(type1,UP+RIGHT*1.2)
        tx2_1 = Tex(r'$H_0\mbox{}$').next_to(tx2,DOWN)
        tx2_2 = Tex(r'$H_1\mbox{}$').next_to(tx2_1,DOWN)
        tx1 = Tex('').next_to(tx2,RIGHT)
        tx1_1 = Tex(r'$\mbox{}H_0$').next_to(tx1,DOWN)
        tx1_2 = Tex(r'$\mbox{}H_0$').next_to(tx1_1,DOWN)
        tx3 = Tex('').next_to(tx1,RIGHT)
        tx3_1 = Tex(r'$\alpha$','=0.05').next_dl(tx3)
        tx3_2 = Tex(r'$\beta$','=').next_dl(tx3_1)
        ani_list = [
            [FadeIn(title1)],
            [FadeIn(type1)],
            [FadeIn(tx2)],
            [FadeIn(tx2_1)],
            [FadeIn(tx1)],
            [FadeIn(tx1_1)],
            [FadeIn(tx3)],
            [FadeIn(tx3_1[0])],
            [FadeIn(tx3_1[1])],
            [FadeIn(type2)],
            [FadeIn(tx2_2)],
            [FadeIn(tx1_2)],
            [FadeIn(tx3_2[0])],
            [FadeIn(tx3_2[1])],
            [FadeOut(title1,type1,type2,tx2,tx2_1,tx2_2,tx1,tx1_1,tx1_2,tx3,tx3_1)],
            [ReplacementTransform(tx3_2,title2)],
        ]
        self.multi_play(ani_list,False)
        return title2
    def politics(self,align_mob):
        tx1 = Tex(r'$\alpha \mbox{} \beta $','').next_dl(align_mob)
        tx2 = Tex('').next_dl(tx1)
        tx3 = Tex(',').next_dl(tx2)
        tx4 = Tex(':').next_dl(tx3)
        tx5 = Tex('').next_dl(tx4)
        tx6 = Tex(',').next_dl(tx5)
        tx7 = Tex('').next_dl(tx6)
        tx8 = Tex('').next_dl(tx7)
        tx9 = Tex('').next_dl(tx8)
        tx10 = Tex('').next_dl(tx9)
        tx2_10 = VGroup(tx2,tx3,tx4,tx5,tx6,tx7,tx8,tx9,tx10)

        tx11 = Tex('',r'$\alpha \mbox{} \beta $','').next_dl(tx1)
        tx12 = Tex(r'$\delta \uparrow$').next_dl(tx11)
        bg12 = BackgroundRectangle(tx12,color=c_red,fill_opacity=0.6)
        tx13 = Tex(r'$\sigma \downarrow$').next_dl(tx12)
        bg13 = BackgroundRectangle(tx13,color=c_red,fill_opacity=0.6)
        tx14 = Tex(r'$n \uparrow$').next_dl(tx13)
        bg14 = BackgroundRectangle(tx14,color=c_green,fill_opacity=0.6)
        tx11_14 = VGroup(tx12,tx13,tx14)
        brace1 = Brace(tx11_14,RIGHT)
        right_arrow = Tex(r'$\rightarrow$').next_to(brace1)

        tx15 = Tex(r'$\alpha \downarrow$')
        tx16 = Tex(r'$\beta \downarrow$').next_dl(tx15)
        tx15_16 = VGroup(tx15,tx16)
        brace2 = Brace(tx15_16,LEFT).next_to(right_arrow)
        tx15_16.next_to(brace2,RIGHT)

        tx17 = Tex('')
        tx18 = Tex('').next_dl(tx17)
        tx19 = Tex('').next_dl(tx18)
        bg17 = BackgroundRectangle(tx17,color=c_red,fill_opacity=0.6)
        bg18 = BackgroundRectangle(tx18,color=c_red,fill_opacity=0.6)
        bg19 = BackgroundRectangle(tx19,color=c_green,fill_opacity=0.6)
        tx17 = VGroup(tx17,bg17)
        tx18 = VGroup(tx18,bg18)
        tx19 = VGroup(tx19,bg19)
        tx17_19 = VGroup(tx17,tx18,tx19).next_to(tx15_16)
        tx20 = Tex('').next_dl(tx14)
        tx21 = Tex(r'$n \uparrow \mbox{}\downarrow ,\mbox{}\uparrow,\mbox{}\uparrow$').next_dl(tx20)
        tx21[0][5:9].set(color=c_green)
        tx21[0][-5:-1].set(color=c_red)
        all_mob = VGroup(tx1,tx11,tx11_14,bg12,bg13,bg14,brace1,right_arrow,tx15_16,brace2,tx17_19,tx20,tx21)

        ani_list = [
            [FadeIn(tx1)],[FadeIn(tx2)],[FadeIn(tx3)],[FadeIn(tx4)],[FadeIn(tx5)],
            [FadeIn(tx6)],[FadeIn(tx7)],[FadeIn(tx8)],[FadeIn(tx9)],[FadeIn(tx10)],
            [FadeOut(tx2_10)],
            [FadeIn(tx11)],[FadeIn(tx12)],[FadeIn(tx13)],[FadeIn(tx14)],
            [FadeIn(brace1)],
            [FadeIn(right_arrow)],
            [FadeIn(brace2)],
            [FadeIn(tx15_16)],
            [FadeIn(tx17),FadeIn(bg12)],
            [FadeIn(tx18),FadeIn(bg13)],
            [FadeIn(tx19),FadeIn(bg14)],
            [FadeIn(tx20)],[FadeIn(tx21)],
            [FadeOut(all_mob)],
        ]
        change_list = [i for i in range(18,22)]
        time_list = self.change_time(len(ani_list),change_list)
        self.multi_play(ani_list,True,time_list)
    def construct(self):
        footnote = foot(c='')
        self.add(footnote)
        self.camera.background_color = c_white
        title_name = Text('',font_size=70,font='STKaiti',color=c_blue)
        title_class = Text('',font_size=40,font='STKaiti',color=c_blue).to_edge(UL)
        department2 = Text('',font_size=40,font='STKaiti',color=c_blue).to_edge(DR)
        department1 = Text('',font_size=40,font='STKaiti',color=c_blue).next_to(department2,UP)
        author = Text('',font_size=40,font='STKaiti',color=c_blue).next_to(department1,UP)        
        self.add(title_name,title_class,author,department1,department2)
        self.wait(2)  
        self.play(FadeOut(title_name,title_class,author,department1,department2))

        cal_beta = self.type_error()

        alpha = 0.05
        n = 25
        mu0 = 0
        mu1 = 1
        sigma = 2
        se = sigma/sqrt(n)       
        boundary_value = norm.isf(alpha,mu0,se) 
        condition1 = Tex('').next_dl(cal_beta[0][1])
        condition2 = Tex(r'$\alpha \mbox{} \beta$','').next_dl(condition1[0][1])
        condition3 = Tex('','','').next_dl(condition2)
        condition3[1].set(color=c_red)
        question = Paragraph('6',
                             '(TBBMC)',
                             'TBBMC2(%)',
                             'TBBMC1(%)',
                             '1(%)',
                             '25(1-)',line_spacing=0.7,
                             t2c={'2':c_red,'1':c_red,'25':c_red,'1(%)':c_green})
        question.next_dl(cal_beta)
        mob_delta = Tex(r'$\delta =1$').next_dl(question)
        mob_std = Tex(r'$\sigma =2$').next_dl(mob_delta)
        mob_n = Tex(r'$n=25$').next_dl(mob_std)
        mob_delta2 = Tex(r'$\delta$','',r'$\mu_{\tiny \mbox{}} - \mu_{\tiny \mbox{}}$').next_dl(mob_n)
        mob_delta3 = Tex(r'$\delta=\mu-\mu_0$','=1').next_dl(mob_n)
        h0_2 = Tex(r'$H_0:\mu=\mu_0\enspace$',r'$\mu-\mu_0=0\enspace$',r'$\mu_0=0$').next_to(mob_std,buff=1)
        h0 = Tex(r'$H_0:$',r'$\mu=0$').next_to(mob_std,buff=1)
        h1_2 = Tex(r'$H_1:\mu > \mu_0\enspace$',r'$\mu-\mu_0=1\enspace$',r'$\mu_0=0$').next_dl(h0)
        h1 = Tex(r'$H_1:$',r'$\mu=1$').next_dl(h0)
        hyper_info_vg = VGroup(mob_std,mob_n,mob_delta)
        h_vg = VGroup(h0,h1)
        h0_true = Tex(r'$H_0$','I').next_dl(cal_beta)
        h1_true = Tex(r'$H_1$','II').next_dl(h0_true)

        ani_list = [
            [FadeIn(condition1)],
            [FadeIn(condition2)],
            [FadeIn(condition3)],
            [FadeOut(condition1,condition2,condition3[0],condition3[2]),condition3[1].animate.next_to(cal_beta)],
            [FadeIn(question)],
            [TransformFromCopy(question[2][-9:-4],mob_std,run_time=2)],
            [TransformFromCopy(question[5][3:5],mob_n,run_time=2)],
            [FadeIn(mob_delta2[0])],
            [FadeIn(mob_delta2[1])],
            [FadeIn(mob_delta2[2])],
            [ReplacementTransform(mob_delta2,mob_delta3[0])],
            [TransformFromCopy(question[4][11:15],mob_delta3[1],run_time=2)],
            [TransformMatchingShapes(mob_delta3,mob_delta,run_time=2)],
            [FadeIn(h0_2[0])],
            [FadeIn(h0_2[1])],
            [FadeIn(h0_2[2])],
            [TransformMatchingShapes(h0_2,h0)],
            [FadeIn(h1_2[0])],
            [FadeIn(h1_2[1])],
            [FadeIn(h1_2[2])],
            [TransformMatchingShapes(h1_2,h1)],
            [FadeOut(question)],
            [hyper_info_vg.animate.next_dl(h1_true),h_vg.animate.next_dl(cal_beta)],
        ]
        # change_list = [i for i in range(3,13)]+[16,20,21,22]
        # time_list = self.change_time(len(ani_list),change_list)
        self.multi_play(ani_list,True)

        mob_alpha = Tex(r'$\alpha=0.05$').next_dl(mob_n)
        mob_se = MathTex(r'\sigma_{\bar x} = \frac{\sigma}{\sqrt{n}} = 0.4').next_to(mob_alpha,DOWN,aligned_edge=LEFT,buff=mob_alpha.get_height()+0.5)
        mob_beta_condition = VGroup(mob_delta,mob_std,mob_n,mob_alpha,mob_se)
        distribution_of_xbar_on_h0 = Tex(r'$\bar X \sim N(0,0.4^2)$').next_to(h0)
        distribution_of_xbar_on_h1 = Tex(r'$\bar X \sim N(1,0.4^2)$').next_to(h1)
        cal_step1 = Tex('1.',r'$H_0:$','',r'$H_0$','').next_dl(cal_beta)
        cal_step1[1].set(color=c_green)
        cal_step1[4][-2:].set(color=MAROON)
        cal_step2 = Tex('2.',r'$H_1:$','',r'$H_0$','').next_dl(cal_step1)
        cal_step2[1].set(color=c_red)
        cal_step2[4][-2:].set(color=c_blue)
        power = Tex('',r'$H_1$','',r'$H_0$','').next_dl(mob_se)
        power[0].set(color=c_green)
                                                                                                                                                                             
        ax = Axes(x_range=[-2,3,1],y_range=[0,1,0.1],x_length=10,y_length=3,
                  x_axis_config={'include_numbers':True},
                  tips=False).set(color=c_black).scale(0.85).to_edge(UR)
        ax.remove(ax[1])
        curve = ax.plot(lambda x: norm.pdf(x,0,0.4),color=c_green)
        area = ax.get_area(curve,[boundary_value,3],color=MAROON_E,opacity=0.7)
        ax3 = ax.copy()
        curve3 = ax3.plot(lambda x: norm.pdf(x,1,0.4),color=c_red)
        area3 = ax.get_area(curve3,[-2,boundary_value],color=c_blue,opacity=0.7)

        line_mu0 = DashedLine(ax.c2p(0,norm.pdf(0,mu0,se)),ax.c2p(0,0),color=c_green)
        line_mu1 = DashedLine(ax3.c2p(1,norm.pdf(1,mu1,se)),ax3.c2p(1,0),color=c_red)
        mu0_value = Tex('0').next_to(line_mu0,DOWN)
        mu0_value2 = mu0_value.copy()
        h0_curve_vg = VGroup(ax,curve,line_mu0,mu0_value)
        vg_3 = VGroup(ax3,curve3,line_mu1,mu0_value2)
        ani_list = [       
            [FadeIn(mob_alpha)],
            [FadeIn(mob_se)],
            [FadeIn(distribution_of_xbar_on_h0)],
            [Create(h0_curve_vg)],
            [distribution_of_xbar_on_h0.animate.align_to(ax,LEFT)],
            [h0.animate.next_to(line_mu0,UP,buff=0.1)],
            [FadeIn(h0_true),FadeOut(distribution_of_xbar_on_h0)],
            [ReplacementTransform(h0_true,area),FadeIn(distribution_of_xbar_on_h0)],
            [FadeIn(distribution_of_xbar_on_h1)],
            [Create(vg_3)],
            [distribution_of_xbar_on_h1.animate.align_to(ax,LEFT)],
            [h1.animate.next_to(line_mu1,UP,buff=0.1)],
            [FadeIn(h1_true),FadeOut(distribution_of_xbar_on_h1)],
            [ReplacementTransform(h1_true,area3),FadeIn(distribution_of_xbar_on_h1)],
        ]
        self.multi_play(ani_list,True)

        distribution_of_xbar_on_h1.move_to(distribution_of_xbar_on_h0)
        vg_3 = VGroup(vg_3,area3,h1,distribution_of_xbar_on_h1)
        ani_list = [
            [vg_3.animate(run_time=2).next_to(ax,DOWN,buff=0.5)],
            ['r',curve,area,curve3,line_mu1,area3,ax3],
        ]
        self.multi_play(ani_list,False)

        tracker_alpha = ValueTracker(0.05)
        tracker_delta = ValueTracker(1)
        tracker_sigma = ValueTracker(2)
        tracker_n     = ValueTracker(25)
        def create_curve():
            n = tracker_n.get_value()
            sigma = tracker_sigma.get_value()
            mu1 = tracker_delta.get_value()
            alpha = tracker_alpha.get_value()
            se = sigma/sqrt(n)
            mu0 = 0
            boundary_value = norm.isf(alpha,mu0,se) 
            curve = ax.plot(lambda x: norm.pdf(x,mu0,se),x_range=[-2,3]).set(color=c_green)
            reject_arrow = Arrow(ax.c2p(boundary_value,0.9),ax.c2p(2,0.9),color=MAROON_E)
            reject_line = DashedLine(ax.c2p(boundary_value,0),ax.c2p(boundary_value,1),color=ORANGE)
            reject_text = Tex(r'$\mbox{}H_0$').next_to(reject_arrow).set(color=MAROON_E)
            non_reject_arrow = Arrow(ax.c2p(boundary_value,0.9),ax.c2p(-1,0.9),color=c_blue)
            non_reject_text = Tex(r'$\mbox{}H_0$').next_to(non_reject_arrow,LEFT).set(color=c_blue)
            area = ax.get_area(curve,x_range=[boundary_value,3],color=MAROON_E,opacity=0.7)            

            ax2 = Axes(x_range=[-2,3,1],y_range=[0,1,0.1],x_length=10,y_length=3,
                  x_axis_config={'include_numbers':True},
                  tips=False).set(color=c_black).scale(0.85).move_to(ax3,aligned_edge=DOWN)
            ax2.remove(ax2[1])
            curve2 = ax2.plot(lambda x:norm.pdf(x,mu1,se),[-2,3]).set(color=c_red)
            line_mu1 = DashedLine(ax2.c2p(mu1,norm.pdf(mu1,mu1,se)),ax2.c2p(mu1,0),color=c_red)
            reject_line2 = DashedLine(ax2.c2p(boundary_value,0),ax2.c2p(boundary_value,1),color=ORANGE)
            h1.next_to(line_mu1,UP,buff=0.1)
            area2 = ax2.get_area(curve2,x_range=[-2,boundary_value],color=c_blue,opacity=0.7)
            line_tm = Line(ax2.c2p(0,0),ax2.c2p(mu1,0)).set(color=c_red)
            brace = Brace(line_tm,direction=DOWN,buff=0)

            curve_h0 = VGroup(curve,area,ax,line_mu0,mu0_value,h0)
            curve_h1 = VGroup(ax2,curve2,area2,line_mu1,h1,mu0_value2)
            reject = VGroup(reject_line,reject_arrow,reject_text,non_reject_arrow,non_reject_text,reject_line2)

            alpha_value = VGroup(Tex(r'$\alpha=$'),DecimalNumber(alpha,num_decimal_places=3)).arrange().set(color=MAROON_E)
            alpha_value.next_to(reject_line,RIGHT*1.5).shift(DOWN*0.7)
            beta_value = VGroup(Tex(r'$\beta=$'),DecimalNumber(norm.cdf(boundary_value,mu1,se),num_decimal_places=3)).arrange().set(color=c_blue)
            beta_value.move_to(ax2.c2p(-1,0.2))
            bound1 = DecimalNumber(boundary_value,num_decimal_places=3,color=ORANGE).scale(0.8).next_to(reject_line,DOWN,buff=0.2)
            bound2 = DecimalNumber(boundary_value,num_decimal_places=3,color=ORANGE).scale(0.8).next_to(ax2.c2p(boundary_value,0),DOWN,buff=0.2)
            text = Tex('',r'$\beta$','').to_edge(RIGHT*1.3)
            text[1].set(color=c_blue)
            h0_distribution1 = Tex(r'$\bar X \sim N(0,$')
            h0_distribution2 = DecimalNumber(se,num_decimal_places=1).next_to(h0_distribution1,buff=0.1)
            h0_distribution3 = Tex(r'$1^2)$').next_to(h0_distribution2,buff=0.01).shift(LEFT*0.17)
            h0_distribution3[0].remove(h0_distribution3[0][0])
            h0_distribution = VGroup(h0_distribution1,h0_distribution2,h0_distribution3).move_to(distribution_of_xbar_on_h0)
            h1_distribution1 = Tex(r'$\bar X \sim N($')
            h1_distribution2 = DecimalNumber(mu1,num_decimal_places=1).next_to(h1_distribution1,buff=0.1)
            h1_distribution3 = Tex(',').next_to(h1_distribution2,buff=0.1,aligned_edge=DOWN)
            h1_distribution4 = DecimalNumber(se,num_decimal_places=1).next_to(h1_distribution3,buff=0.1,aligned_edge=DOWN)
            h1_distribution5 = Tex(r'$1^2)$').next_to(h1_distribution4,buff=0.01).shift(LEFT*0.17)
            h1_distribution5[0].remove(h1_distribution5[0][0])
            h1_distribution = VGroup(h1_distribution1,h1_distribution2,h1_distribution3,h1_distribution4,h1_distribution5).move_to(distribution_of_xbar_on_h1)
            alpha_beta = VGroup(alpha_value,beta_value,bound1,bound2,brace,text,h0_distribution,h1_distribution)

            vg = VGroup(curve_h0,curve_h1,reject,alpha_beta)
            return vg
        
        def curve_delta():
            h1 = VGroup(Tex(r'$H_1:\mu=$'),DecimalNumber(tracker_delta.get_value(),num_decimal_places=1)).arrange()
            vg = create_curve()
            factor2 = Tex(r'$\delta=1 \sim 0.5$').next_dl(vg[3][5])
            vg[1].remove(vg[1][4])
            factor = VGroup(Tex(r'$\delta=$'),DecimalNumber(tracker_delta.get_value(),num_decimal_places=3)).arrange()
            factor.next_to(factor2,DOWN)
            h1.next_to(vg[1][3],UP,buff=0.1)
            vg = VGroup(vg,factor2,factor,h1)
            return vg
        def curve_sigma():
            vg = create_curve()
            factor2 = Tex(r'$\sigma=2 \sim 5$').next_dl(vg[3][5])
            factor = VGroup(Tex(r'$\sigma=$'),DecimalNumber(tracker_sigma.get_value(),num_decimal_places=3)).arrange()
            factor.next_to(factor2,DOWN)
            sigma_x = tracker_sigma.get_value()/sqrt(tracker_n.get_value())
            factor3 = MathTex(r'\sigma_{\bar x} &= \sigma / \sqrt{n}\\ &=').next_dl(factor)
            factor3_numb = DecimalNumber(sigma_x,num_decimal_places=3).next_to(factor3[0][-1])
            factor3 = VGroup(factor3,factor3_numb)
            vg[3].remove(vg[3][4])
            vg = VGroup(vg,factor2,factor,factor3)
            return vg
        def curve_n():
            vg = create_curve()
            factor2 = Tex(r'$n=25 \sim 50$').next_dl(vg[3][5])
            factor = VGroup(Tex(r'$n=$'),DecimalNumber(tracker_n.get_value(),num_decimal_places=0)).arrange()
            factor.next_to(factor2,DOWN)
            sigma_x = tracker_sigma.get_value()/sqrt(tracker_n.get_value())
            factor3 = MathTex(r'\sigma_{\bar x} &= \sigma / \sqrt{n}\\ &=').next_dl(factor)
            factor3_numb = DecimalNumber(sigma_x,num_decimal_places=3).next_to(factor3[0][-1])
            factor3 = VGroup(factor3,factor3_numb)            
            vg[3].remove(vg[3][4])
            vg = VGroup(vg,factor2,factor,factor3)
            return vg
        def curve_alpha():
            vg = create_curve()
            factor2 = Tex(r'$\alpha=0.05 \sim 0.01$').next_dl(vg[3][5])
            factor = VGroup(Tex(r'$\alpha=$'),DecimalNumber(tracker_alpha.get_value(),num_decimal_places=3)).arrange()
            factor.next_to(factor2,DOWN)
            vg[3].remove(vg[3][4])
            vg = VGroup(vg,factor2,factor)
            return vg
        def play_curve(curve,tracker,value1,value2,runtime=3):
            self.add(curve)
            self.wait()
            self.play(tracker.animate(run_time=runtime).set_value(value1))
            self.wait()
            self.play(tracker.animate(run_time=runtime).set_value(value2))
            self.wait()
            self.remove(curve)

        def play_delta():
            curve_area = always_redraw(curve_delta)
            cal_alpha_1 = Tex(r'$Z_{0.05}=1.645$').next_to(curve_area[0][3][0],UP,aligned_edge=LEFT)
            cal_alpha_2 = Tex(r'$\bar X_{0.05}=\mu_0+Z_{0.05}\times \sigma_{\bar x}$').next_to(cal_alpha_1,UP,aligned_edge=LEFT)
            cal_alpha_3 = Tex(r'$\bar X_{0.05}=0+1.645 \times 0.4$').next_to(cal_alpha_1,UP,aligned_edge=LEFT)
            cal_alpha_4 = Tex(r'$\bar X_{0.05}=0.658$').next_to(cal_alpha_1,UP,aligned_edge=LEFT)
            cal_beta_1 = MathTex(r'Pr(\bar X<=0.658)').move_to(curve_area[0][3][1])
            cal_beta_2 = MathTex(r'Z = \frac{(\bar x-\mu)}{\sigma_{\bar x}}').move_to(curve_area[0][3][1])
            cal_beta_3 = MathTex(r'Z = \frac{(0.658-1)}{0.4}').move_to(curve_area[0][3][1])
            cal_beta_4 = MathTex(r'Z = -0.855').move_to(curve_area[0][3][1])
            cal_beta_5 = MathTex(r'Pr(\bar X<=0.658)=0.196').move_to(curve_area[0][3][1])
            power_name = Tex('').set(color=c_green).move_to(curve_area[1])
            power_name2 = Tex(r'($1-\beta$)').next_to(power_name,DOWN)
            power_area = curve_area[0][1][0].get_area(curve_area[0][1][1],[0.658,3],color=c_green,opacity=0.7)

            self.add(curve_area[0][0],curve_area[0][1])
            ani_list = [       
                [FadeIn(curve_area[0][3][0])],
                [FadeIn(cal_step1)],
                [FadeIn(cal_alpha_1)],
                [FadeIn(cal_alpha_2)],
                [TransformMatchingShapes(cal_alpha_2,cal_alpha_3)],
                [TransformMatchingShapes(cal_alpha_3,cal_alpha_4)],
                [ReplacementTransform(cal_alpha_4,curve_area[0][3][2]),FadeOut(cal_alpha_1)],
                [FadeIn(curve_area[0][2][0])],
                [FadeIn(curve_area[0][2][1])],
                [FadeIn(curve_area[0][2][2])],
                [FadeIn(curve_area[0][2][3])],
                [FadeIn(curve_area[0][2][4])],                
                [FadeIn(curve_area[0][2][5])],
                [FadeIn(curve_area[0][3][3])],
                [FadeIn(power_area)],
                [FadeIn(power_name2)],
                [FadeIn(power_name)],
                [FadeIn(power)],
                [FadeIn(cal_step2),FadeOut(power_area,power_name2,power_name,power)],
                [FadeIn(cal_beta_1)],
                [TransformMatchingShapes(cal_beta_1,cal_beta_2)],
                [TransformMatchingShapes(cal_beta_2,cal_beta_3)],
                [TransformMatchingShapes(cal_beta_3,cal_beta_4)],
                [TransformMatchingShapes(cal_beta_4,cal_beta_5)],
                [ReplacementTransform(cal_beta_5,curve_area[0][3][1])],
                [FadeIn(curve_area[0][3][5])],
                [FadeIn(curve_area[1])],
                [FadeIn(curve_area[2])],
            ]
            self.multi_play(ani_list,True)
            self.remove(distribution_of_xbar_on_h0,distribution_of_xbar_on_h1)
            self.remove(h1)
            play_curve(curve_area,tracker_delta,0.5,1)

        def play_sigma():
            curve_area = always_redraw(curve_sigma)
            self.add(curve_area)
            self.remove(curve_area[1],curve_area[2],curve_area[3])
            ani_list = [       
                [FadeIn(curve_area[1])],
                [FadeIn(curve_area[2])],
                [FadeIn(curve_area[3])],
            ]
            self.multi_play(ani_list)
            play_curve(curve_area,tracker_sigma,5,2)

        def play_n():
            curve_area = always_redraw(curve_n)
            self.add(curve_area)
            self.remove(curve_area[1],curve_area[2],curve_area[3])
            ani_list = [       
                [FadeIn(curve_area[1])],
                [FadeIn(curve_area[2])],
                [FadeIn(curve_area[3])],
            ]
            self.multi_play(ani_list)
            play_curve(curve_area,tracker_n,50,25)

        def play_alpha():
            curve_area = always_redraw(curve_alpha)
            self.add(curve_area)
            self.remove(curve_area[1],curve_area[2])
            ani_list = [       
                [FadeIn(curve_area[1])],
                [FadeIn(curve_area[2])],
            ]
            self.multi_play(ani_list)
            play_curve(curve_area,tracker_alpha,0.01,0.05)
            return curve_area

        play_delta()
        play_sigma()
        play_n()
        curve_area = play_alpha()

        beta_condition = VGroup(mob_delta,mob_std,mob_n,mob_alpha)
        brace_condition = Brace(beta_condition,RIGHT)

        ani_list=[
            ['r',cal_step1,cal_step2],
            [FadeIn(brace_condition)],
            [condition3[1].animate.next_to(brace_condition)],
        ]
        time_list = [0,1,1]
        self.multi_play(ani_list,True,time_list)
        beta_value = Tex(r'$\beta=0.196$').set(color=c_blue).move_to(condition3[1])
        self.play(ReplacementTransform(condition3[1],beta_value))
        self.wait()
        mob_beta_condition.add(brace_condition,beta_value)

        self.play(FadeOut(mob_beta_condition))
        self.politics(cal_beta[0][1])

        t1 = Tex('').next_dl(cal_beta[0][1])
        power = Tex('','',r'$H_1$','',r'$H_0$','').next_dl(t1[0][1])
        power[0].set(color=c_green)
        t2 = Tex('').next_to(power,DOWN,buff=0.5).align_to(cal_beta[0][1],LEFT)
        # step1 = Tex('',r'$H_0$','',r'$\alpha=0.05$','').next_dl(t2[0][1])
        step1 = Tex('',r'$H_0$','',r'$\bar x$','').next_dl(t2[0][1])
        # step2 = Tex('',r'$H_1$','',r'$\bar x$','').next_dl(step1)
        step2 = Tex('',r'$H_1$','',r'$H_1$','').next_dl(step1)
        t3 = Tex('').next_to(step2,DOWN,buff=0.5).align_to(cal_beta[0][1],LEFT)
        cond = Tex(r'$\delta,\sigma,n,\alpha$').next_dl(t3[0][1])
        ani_list = [
            [FadeIn(t1)],
            [FadeIn(power)],
            [FadeIn(t2)],
            [FadeIn(step1)],
            [FadeIn(step2)],
            [FadeIn(t3)],
            [FadeIn(cond)],
        ]
        self.multi_play(ani_list,True)

        
        
        self.wait(2)
        
```
