```python
from manim import *
```

```python
%%manim -ql Code
class Code(Scene):
    def construct(self):
        m = MathTex(r'\mathbb{M}',fill_color=BLUE,color=RED,fill_opacity=1).scale(7).shift(UP+LEFT)
        c = Circle(fill_color=GREEN,fill_opacity=0.8,stroke_width=0).shift(DOWN)
        s = Square(fill_color=RED_E,fill_opacity=0.9,stroke_width=0).shift(RIGHT)
        t = Triangle(fill_color=YELLOW,stroke_width=0,fill_opacity=0.9).shift(RIGHT+DOWN)
        logo = VGroup(s,c,t,m)
        self.add(logo)
```
